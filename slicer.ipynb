```python
# import dependencies 

import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
from csv import reader
```

```python
opened_file = open('data/dk_data.csv')
read_file = reader(opened_file)
dk = list(read_file)
dk_header = dk[0]
dk = dk[1:]
```

```python
# creating the main function 

def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
        
print(dk_header)
print('\n')
explore_data(dk, 0, 2, True)
```

```python
# create a freq table to build a percentage table


def freq_table(dataset, index):
    table = {}
    total = 0 
    
    for row in dataset:
        total += 1 
        value = row[index]
        if value in table:
            table[value] += 1 
        else:
            table[value] = 1 
            
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage
        
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse=True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
# isolate all accounts for top five accounts with most transHilux, LCM, PoluxMetastar

dk_metastar = []
dk_lcm = []
dk_hilux = []
dk_polux = []
dk_baktelekom = []
dk_faberlex = []
dk_jetfield = []
dk_riverlane = []
dk_skn = []

for s in dk:
    payer_name_norm = s[5]
    if payer_name_norm == 'METASTAR INVEST LLP':
        dk_metastar.append(s)
    if payer_name_norm == 'LCM ALLIANCE LLP':
        dk_lcm.append(s)
    if payer_name_norm == 'HILUX SERVICES LP':
        dk_hilux.append(s)
    if payer_name_norm == 'POLUX MANAGEMENT LP':
        dk_polux.append(s)
    if payer_name_norm == 'BAKTELEKOM MMC':
        dk_baktelekom.append(s)
    if payer_name_norm == 'FABERLEX LP':
        dk_faberlex.append(s)
    if payer_name_norm == 'JETFIELD NETWORKS LIMITED':
        dk_jetfield.append(s)
    if payer_name_norm == 'RIVERLANE LLP':
        dk_riverlane.append(s)
    if payer_name_norm == 'SKN ELECTRICAL SERVICES LIMITED':
        dk_skn.append(s)
        
print(dk_header)
print('\n')
explore_data(dk_metastar, 0, 1, True)
print('\n')
explore_data(dk_lcm, 0, 1, True)
print('\n')
explore_data(dk_hilux, 0, 1, True)
print('\n')
explore_data(dk_polux, 0, 1, True)
print('\n')
explore_data(dk_baktelekom, 0, 1, True)
print('\n')
explore_data(dk_faberlex, 0, 1, True)
print('\n')
explore_data(dk_jetfield, 0, 1, True)
print('\n')
explore_data(dk_riverlane, 0, 1, True)
print('\n')
explore_data(dk_skn, 0, 1, True)
```

```python
# create a count function

def count(dataset, index):
    table = {}
    total = 0 
    
    for row in dataset:
        total += 1 
        value = row[index]
        if value in table:
            table[value] += 1 
        else:
            table[value] = 1 
    return table
```
