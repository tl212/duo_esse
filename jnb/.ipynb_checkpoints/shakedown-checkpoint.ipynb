```python
# dk pandas

dk_df = pd.read_csv('data/dk_data.csv')
dk_df.head()
```

```python
dk_df.shape
```

```python
# sum total ammount by payers_name_norm 

contributors = dk_df.groupby('payer_name_norm')['amount_usd'].sum()
contributors.head(20)
```

```python
contributors.dtype
```

```python
contributors = contributors.to_frame()
contributors.head()
```

```python
contributors = contributors.sort_values('amount_usd', ascending=False)
```

```python
contributors.head()
```

```python
contributors_minus4 = contributors.drop(['LCM ALLIANCE LLP', 'METASTAR INVEST LLP', 'POLUX MANAGEMENT LP', 'HILUX SERVICES LP'])
```

```python
contributors_minus4
```
